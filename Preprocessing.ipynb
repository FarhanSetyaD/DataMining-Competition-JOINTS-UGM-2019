```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test_data.csv')
train = train.drop("id", axis=1)
train
# test
```

```python
train = train.dropna()
train = train.reset_index(drop=True)
train
```

```python
dictionaries = {'putri': -1, 'putra': 1, 'campur': 0}
df_gender = train['gender']
df_gender
# df_gender
#dictionaries['putri']
```

```python
number = []
for i in df_gender:
    
    numbered = dictionaries[i]
    number.append(numbered)
# number
```

```python
dict_data = {'gender': number}
df_gender = pd.DataFrame.from_dict(dict_data)
# df_gender
```

```python
train['gender'] = df_gender['gender']
train
```

```python
# train.corr()
# train.describe()
```

```python
test_id = test['id']
test = test.drop("id", axis=1)
# test_id
```

```python
train_fitur=train.drop('gender',axis=1)
# train_fitur
train_label = train['gender']
# train_gender
X_train, X_test, Y_train, Y_test = train_test_split(train_fitur, train_label)
```

```python
rf = RandomForestClassifier(n_estimators = 1000, random_state = 42)
rf.fit(train_fitur,train_label)
# rf.score(train_fitur), train_label)
predict = rf.predict(test)
predict
```

```python
type(test_id)
converter = {-1: 'putri', 0:'campur', 1:'putra'}

predict = [converter[x] for x in predict]
predict_series = pd.Series(data=predict, name='gender')
submission = pd.DataFrame([test_id, predict_series])
submission = submission.transpose()
submission.to_csv('first_submission_randomforest.csv', index=False)
```
