```python
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('data/train.csv')
train = train.drop('id', axis=1)
train = train.drop('fac_7', axis=1)
train = train.fillna(train.mean())
train['gender']=train['gender'].map({'putri': 1, 'putra': 2, 'campur': 3})
test = pd.read_csv('data/test_data.csv')
len(train)
```

```python
train_fitur = train.drop('gender', axis=1)
train_fitur
```

```python
# dictionaries = {'putri': 1, 'putra': 2, 'campur': 3}
train_label = train['gender']
train_label
```

```python
fac1 = tf.feature_column.numeric_column('fac_1')
fac2 = tf.feature_column.numeric_column('fac_2')
fac3 = tf.feature_column.numeric_column('fac_3')
fac4 = tf.feature_column.numeric_column('fac_4')
fac5 = tf.feature_column.numeric_column('fac_5')
fac6 = tf.feature_column.numeric_column('fac_6')
# fac7 = tf.feature_column.numeric_column('fac_7')
fac8 = tf.feature_column.numeric_column('fac_8')
poi1 = tf.feature_column.numeric_column('poi_1')
poi2 = tf.feature_column.numeric_column('poi_2')
poi3 = tf.feature_column.numeric_column('poi_3')
size = tf.feature_column.numeric_column('size')
price = tf.feature_column.numeric_column('price_monthly')
room = tf.feature_column.numeric_column('room_count')
call = tf.feature_column.numeric_column('total_call')
```

```python
features = [fac1,fac2,fac3,fac4,fac5,fac6,fac8,poi1,poi2,poi3,size,price,room,call]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(train_fitur, train_label, test_size=0.3, random_state=101)
```

```python
model=tf.estimator.BoostedTreesClassifier(feature_columns=features,n_classes=4, n_batches_per_layer=100)
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=train_fitur,y=train_label,batch_size=3293,num_epochs=1,shuffle=True)
model.train(input_fn=input_func, steps=1000)
```

```python
pred = tf.estimator.inputs.pandas_input_fn(x=test, batch_size=824, num_epochs=1, shuffle=False)
prediction = model.predict(input_fn=pred)
final=[]
for i in prediction:
    final.append(i['class_ids'][0])
pd.DataFrame(final)
```
