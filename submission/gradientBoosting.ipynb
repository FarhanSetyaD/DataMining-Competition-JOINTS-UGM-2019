```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, scale
import math
```

```python
train = pd.read_csv('data/train_num.csv')
test = pd.read_csv('data/test_data.csv')
train = train.drop(columns=["id","fac_7"])
train
```

```python
one_null = pd.DataFrame(columns=train.columns)

for i, rows in train.iterrows():
    cond = sum([1 if math.isnan(x) else 0 for x in rows ])
    if (cond ==1):
        one_null = one_null.append(rows)
one_null
```

```python
no_na_train = train.dropna()
#no_na_train.to_csv('no_na_train.csv', index=False)
imputers = {
    'fac_1': KNeighborsClassifier(n_neighbors = 17), 'fac_2': KNeighborsClassifier(n_neighbors = 17), 
    'fac_3': KNeighborsClassifier(n_neighbors = 17), 'fac_4': KNeighborsClassifier(n_neighbors = 17), 
    'fac_5': KNeighborsClassifier(n_neighbors = 17), 'fac_6': KNeighborsClassifier(n_neighbors = 17),
    'fac_8': KNeighborsClassifier(n_neighbors = 17),
    'poi_1': LinearRegression(), 'poi_2': LinearRegression(), 'poi_3': LinearRegression(), 
    'size': LinearRegression(),'price_monthly': LinearRegression(),'room_count': LinearRegression()
}
scaler = {
    'fac_1': StandardScaler(), 'fac_2': StandardScaler(), 
    'fac_3': StandardScaler(), 'fac_4': StandardScaler(), 
    'fac_5': StandardScaler(), 'fac_6': StandardScaler(),
    'fac_8': StandardScaler(),
}

for key in imputers:
    if( type(imputers[key]) == KNeighborsClassifier):
        data = no_na_train.drop(columns=key).astype(np.float64)
        scaler[key].fit(data)
        imputers[key].fit(scaler[key].transform(data), no_na_train[key])
    else:
        imputers[key].fit(no_na_train.drop(columns=key), no_na_train[key])
"success"
```

```python
missing_predict = pd.DataFrame(columns=train.columns)
for i, row in one_null.iterrows():
    row = row.copy()
    key = ""
    for j in imputers:
        if (math.isnan(row[j])):
            key = j
            break
    
    if( type(imputers[key]) == KNeighborsClassifier):
        pre_data = scaler[key].transform([row.drop(key).astype(np.float64)])
        predict = imputers[key].predict(pre_data)
        #print(predict)
        row[key] = predict[0]
        missing_predict = missing_predict.append(row)
    else:
        pre_data = [row.drop(key)]
        predict = imputers[key].predict(pre_data)
        #print(predict)
        row[key] = predict[0]
        missing_predict = missing_predict.append(row)
    
missing_predict
```

```python
train_fill_predict = pd.concat([no_na_train, missing_predict], ignore_index = True)
train_fill_predict
```

```python
x_train, x_test, y_train, y_test = train_test_split(train_fill_predict.drop(columns="gender"), train_fill_predict['gender'])
```

```python
for i in range(5):
    gb = GradientBoostingClassifier(n_estimators=2900, max_depth=8)
    gb.fit(x_train, y_train)
    print(gb.score(x_test, y_test))
```

```python
gb = GradientBoostingClassifier(n_estimators=2900, max_depth=8)
gb.fit(train_fill_predict.drop(columns="gender"), train_fill_predict['gender'])
predict = gb.predict(test.drop(columns=['id', 'fac_7']))
```

```python
id_test = test['id']
submission = pd.DataFrame(columns=['id','gender'])
map_to_gender = {-1.:'putri', 0.: 'campur', 1.:'putra'}

i = 0
for row in id_test:
    submission = submission.append({'id': row, 'gender':map_to_gender[predict[i]]}, ignore_index=True)
    i = i+1
submission
```
